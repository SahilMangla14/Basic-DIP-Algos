```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image, ImageChops
import PIL
import math, operator
```

```python
img = cv2.imread('./images/img1.jpg')
```

```python
def bit_plane_slicing(img):
    planes = []
    for i in range(8):
        plane = np.zeros(img.shape, np.uint8)
        plane[np.where(img & (1 << i))] = 255
        planes.append(plane)
    return planes
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
planes = bit_plane_slicing(gray_image)
```

```python
images = [gray_image,planes[0],planes[1],planes[2],planes[3],planes[4],planes[5],planes[6],planes[7]]
plt.figure(figsize=[9,9])
plt.subplot(3,3,1); plt.imshow(images[0],cmap='gray'); plt.title('Original')
plt.subplot(3,3,2); plt.imshow(images[1],cmap='gray'); plt.title('0th bit')
plt.subplot(3,3,3); plt.imshow(images[2],cmap='gray'); plt.title('1st bit')
plt.subplot(3,3,4); plt.imshow(images[3],cmap='gray'); plt.title('2nd bit')
plt.subplot(3,3,5); plt.imshow(images[4],cmap='gray'); plt.title('3rd bit')
plt.subplot(3,3,6); plt.imshow(images[5],cmap='gray'); plt.title('4th bit')
plt.subplot(3,3,7); plt.imshow(images[6],cmap='gray'); plt.title('5th bit')
plt.subplot(3,3,8); plt.imshow(images[7],cmap='gray'); plt.title('6th bit')
plt.subplot(3,3,9); plt.imshow(images[8],cmap='gray'); plt.title('7th bit')
```
